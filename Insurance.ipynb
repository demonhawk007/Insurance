```python
dataset = concat_df[concat_df['Label'] == 'train']
dataset2 = concat_df[concat_df['Label'] == 'test']
```

```python
dataset = dataset.drop('Label', axis=1)
dataset2 = dataset2.drop('Label', axis=1)
```

```python
x=dataset.iloc[:,0:85].values
y=dataset.iloc[:,-1].values
```

```python
y
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression(random_state=1,class_weight='balanced')
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
#creating Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```
